```python
import torch
import torch.nn as nn
import torch.optim as optim
import torchvision
import torchvision.transforms as transforms
from torch.utils.data import DataLoader
import matplotlib.pyplot as plt

# Set the device for computation (GPU if available, otherwise CPU)
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# Hyperparameters
latent_size = 64          # Size of the noise vector (input to the Generator)
hidden_size = 256         # Number of neurons in the hidden layers of both networks
image_size = 784          # MNIST images are 28x28 pixels, flattened to a 784-dimensional vector
batch_size = 100          # Number of images processed in one forward/backward pass
learning_rate = 0.0002    # Step size for the optimizer
num_epochs = 200          # Number of times the entire dataset is passed through the networks during training

# Data transformations (convert to tensor and normalize to [-1, 1])
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize(mean=(0.5,), std=(0.5,))
])

# Load the MNIST dataset
mnist = torchvision.datasets.MNIST(root='./data', train=True, transform=transform, download=True)
data_loader = DataLoader(dataset=mnist, batch_size=batch_size, shuffle=True)

# Generator network
class Generator(nn.Module):
    def __init__(self, input_size, hidden_size, output_size):
        super(Generator, self).__init__()
        # Fully connected layers
        self.fc1 = nn.Linear(input_size, hidden_size)
        self.fc2 = nn.Linear(hidden_size, hidden_size)
        self.fc3 = nn.Linear(hidden_size, output_size)
        # Activation functions
        self.relu = nn.ReLU()
        self.tanh = nn.Tanh()

    def forward(self, x):
        # Pass through the first hidden layer and apply ReLU activation
        x = self.relu(self.fc1(x))
        # Pass through the second hidden layer and apply ReLU activation
        x = self.relu(self.fc2(x))
        # Pass through the output layer and apply Tanh activation to match the image data range [-1, 1]
        x = self.tanh(self.fc3(x))
        return x

# Discriminator network
class Discriminator(nn.Module):
    def __init__(self, input_size, hidden_size):
        super(Discriminator, self).__init__()
        # Fully connected layers
        self.fc1 = nn.Linear(input_size, hidden_size)
        self.fc2 = nn.Linear(hidden_size, hidden_size)
        self.fc3 = nn.Linear(hidden_size, 1)
        # Activation functions
        self.relu = nn.ReLU()
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        # Pass through the first hidden layer and apply ReLU activation
        x = self.relu(self.fc1(x))
        # Pass through the second hidden layer and apply ReLU activation
        x = self.relu(self.fc2(x))
        # Pass through the output layer and apply Sigmoid activation to get a probability
        x = self.sigmoid(self.fc3(x))
        return x

# Instantiate the Generator and Discriminator models
G = Generator(latent_size, hidden_size, image_size).to(device)
D = Discriminator(image_size, hidden_size).to(device)

# Loss function (Binary Cross Entropy Loss)
criterion = nn.BCELoss()

# Optimizers for both Generator and Discriminator
d_optimizer = optim.Adam(D.parameters(), lr=learning_rate)
g_optimizer = optim.Adam(G.parameters(), lr=learning_rate)

# Training loop
for epoch in range(num_epochs):
    for i, (images, _) in enumerate(data_loader):
        # Flatten the MNIST images from [batch_size, 1, 28, 28] to [batch_size, 784]
        images = images.reshape(batch_size, -1).to(device)

        # Create labels for real images (1s) and fake images (0s)
        real_labels = torch.ones(batch_size, 1).to(device)
        fake_labels = torch.zeros(batch_size, 1).to(device)

        # Train Discriminator
        # Forward pass with real images
        outputs = D(images)
        d_loss_real = criterion(outputs, real_labels)  # Real image loss
        real_score = outputs

        # Generate fake images from random noise
        z = torch.randn(batch_size, latent_size).to(device)
        fake_images = G(z)
        outputs = D(fake_images)
        d_loss_fake = criterion(outputs, fake_labels)  # Fake image loss
        fake_score = outputs

        # Total Discriminator loss
        d_loss = d_loss_real + d_loss_fake

        # Backpropagation and optimization for the Discriminator
        d_optimizer.zero_grad()
        d_loss.backward()
        d_optimizer.step()

        # Train Generator
        # Generate new fake images from random noise
        z = torch.randn(batch_size, latent_size).to(device)
        fake_images = G(z)
        outputs = D(fake_images)

        # Generator's loss (trying to fool the Discriminator)
        g_loss = criterion(outputs, real_labels)

        # Backpropagation and optimization for the Generator
        g_optimizer.zero_grad()
        g_loss.backward()
        g_optimizer.step()

        # Print progress every 200 steps
        if (i + 1) % 200 == 0:
            print(f'Epoch [{epoch}/{num_epochs}], Step [{i + 1}/{len(data_loader)}], '
                  f'D Loss: {d_loss.item():.4f}, G Loss: {g_loss.item():.4f}, '
                  f'D(x): {real_score.mean().item():.2f}, D(G(z)): {fake_score.mean().item():.2f}')

    # Save and visualize generated images every 20 epochs
    if epoch  % 20 == 0:
        with torch.no_grad():
            fake_images = fake_images.reshape(fake_images.size(0), 1, 28, 28)
            grid = torchvision.utils.make_grid(fake_images, nrow=10, normalize=True)
            plt.imshow(grid.permute(1, 2, 0).cpu().numpy())
            plt.show()

# Save the trained models
torch.save(G.state_dict(), 'generator.pth')
torch.save(D.state_dict(), 'discriminator.pth')
```

```python
# Visualizing the Generated Images
# To see what the generated images look like, you can generate some images using the Generator and then visualize them. Here's how you can do it:

import torchvision.utils as vutils

# Generate a batch of random noise vectors
z = torch.randn(batch_size, latent_size).to(device)

# Use the Generator to produce fake images from the noise
fake_images = G(z)

# Reshape the fake images from [batch_size, 784] to [batch_size, 1, 28, 28]
fake_images = fake_images.view(fake_images.size(0), 1, 28, 28)

# Create a grid of images for visualization
grid = vutils.make_grid(fake_images, nrow=10, normalize=True)

# Plot the grid of images
plt.figure(figsize=(10, 10))
plt.imshow(grid.permute(1, 2, 0).cpu().numpy())
plt.title("Generated Images")
plt.axis("off")
plt.show()
```

```python
import matplotlib.pyplot as plt
import torchvision.transforms as transforms
import torchvision.datasets as datasets
from torch.utils.data import DataLoader

# Define the transformation for the MNIST dataset (normalize to [-1, 1] and convert to tensor)
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.5,), (0.5,))
])

# Load the MNIST dataset
mnist = datasets.MNIST(root='./data', train=True, transform=transform, download=True)
data_loader = DataLoader(dataset=mnist, batch_size=8, shuffle=True)

# Get a batch of images and their corresponding labels
images, labels = next(iter(data_loader))

# Plotting the images with their labels
plt.figure(figsize=(12, 12))
for i in range(images.shape[0]):
    plt.subplot(4, 4, i + 1)
    # Reshape the image to its original size and denormalize it to [0, 1]
    image = images[i].view(28, 28).numpy()
    image = (image * 0.5) + 0.5  # Denormalize the image
    plt.imshow(image, cmap='gray')
    plt.title(f"Label: {labels[i].item()}")
    plt.axis('off')
plt.show()
```
